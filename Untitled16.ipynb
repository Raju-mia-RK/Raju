```python

students = pd.read_csv('/content/drive/MyDrive/datasets/StudentsPerformance.csv')
```

```python
def edaFunction(df):
  fig,axes=plt.subplots(2,3,figsize=(12,8))
  math_score=df['math score']
  axes[0,0].hist(math_score,bins=20)
  axes[0,0].set_title("Math score")

  gender=df['gender'].value_counts()
  axes[0,1].bar(gender.index,gender.values)
  axes[0,1].set_title("Gender")

  prep_score = df['test preparation course'].value_counts()
  axes[0,2].pie(prep_score.values, labels=prep_score.index, autopct='%1.0f%%')
  axes[0,2].set_title("Test Preparation")

  axes[1,0].scatter(df['math score'],df['reading score'],alpha=0.3)
  axes[1,0].set_title("Math vs Reading")

  lunch_avg=df.groupby('lunch')['math score'].mean()
  axes[1,1].bar(lunch_avg.index,lunch_avg.values)
  axes[1,1].set_title("Lunch Effect")

  avg=[df['math score'].mean(),df['reading score'].mean(),df['writing score'].mean()]
  axes[1,2].bar(['Math','Reading','Writing'],avg)
  axes[1,2].set_title('Average Scores')
  plt.tight_layout()
  plt.show()

edaFunction(students)
```

```python
def gymFunction(df):
  fig,axes=plt.subplots(2,2,figsize=(10,8))
  axes[0,0].hist(gym['Age'],bins=15)
  axes[0,0].set_title('Age Distribution')

  gender=df['gender'].value_counts()
  axes[0,1].bar(gender.index,gender.values)
  axes[0,1].set_title("Gender")

  axes[1,0].hist(gym['visit_per_week'],bins=10)
  axes[1,0].set_title('visits per week')

  pt=df['personal_training'].value_counts()
  axes[1,1].pie(pt.values,labels=pt.index,autopct='%1.0f%%')
  axes[1,1].set_title('Personal Training')

  plt.tight_layout()
  plt.show()

gymFunction(gym)
```
