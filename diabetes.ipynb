```python
x=df.drop('Outcome',axis=1)
y=df['Outcome']

x_train,x_test,y_train,y_test=train_test_split(

  x,y,test_size=0.2,random_state=42
)

print(f"\n we will train with: {len(x_train)} patients")
print(f"We will test with: {len(x_test)} patients")
```

```python
print("Before scaling-different ranges")
print(f"Age: smallest={x_train['Age'].min()}, biggest={x_train['Age'].max()}")
print(f"Glucose: smallest={x_train['Glucose'].min()}, biggest={x_train['Glucose'].max()}")
print(f"Pregnancies: smallest={x_train['Pregnancies'].min()}, biggest={x_train['Pregnancies'].max()}")
```

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
print("After scaling")
print(f"First patients age before scaling: {x_train.values[0][7]}")
print(f"First patients age after scaling: {x_train_scaled[0][7]}")
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax_scaler=MinMaxScaler()
x_train_minmax=minmax_scaler.fit_transform(x_train)
x_test_minmax=minmax_scaler.transform(x_test)
print(x_train_minmax.min())
print(x_train_minmax.max())
print(x_test_minmax.min())
print(x_test_minmax.max())

print(x_train_scaled.min())
print(x_train_scaled.max())
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train_scaled,y_train)
print("check accuracy")
train_score=model.score(x_train_scaled,y_train)
test_score=model.score(x_test_scaled,y_test)
print(f"Train accuracy: {train_score:.1%}")
print(f"Test accuracy: {test_score:.1%}")
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(
    model,x_train_scaled,y_train,cv=5
)
print('5 mini-test scores: ')
for i in range(5):
  print(f"Test accuracy: {scores[i]:.1%}")
print(f'Average: {scores.mean():.1%}')
```

```python
import joblib
joblib.dump(model,'my_model.pkl')
joblib.dump(scaler,'my_scaler.pkl')
pd.DataFrame(x_train_scaled).to_csv('train_data_scaled.csv',index=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import joblib

df1=pd.read_csv('/content/drive/MyDrive/datasets/diabetes.csv')
x=df1.drop('Outcome',axis=1)
y=df1['Outcome']
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.2,random_state=42
)
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)

model=LogisticRegression(max_iter=1000)
model.fit(x_train_scaled,y_train)
print('model trained')
accuracy=model.score(x_test_scaled,y_test)
print(f"Accuracy: {accuracy: .1%}")
joblib.dump(model,'model.pkl')
joblib.dump(scaler,'scaler.pkl')
```
