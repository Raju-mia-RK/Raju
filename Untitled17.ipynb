```python
gym=pd.read_csv('/content/drive/MyDrive/datasets/gym_membership.csv')
```

```python
anonoment_value=gym['abonoment_type'].value_counts()
plt.bar(anonoment_value.index,anonoment_value.values)

plt.tight_layout()
plt.show()
```

```python
plt.hist(gym['avg_time_in_gym'],bins=15)
plt.title('Average time in gym')
plt.tight_layout()
plt.show()
```

```python
plt.scatter(gym['Age'],gym['visit_per_week'])
plt.grid(True,which='both',linestyle='--',linewidth=0.7,alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=gym,x='gender',y='Age')
plt.grid(True,which='both',linestyle='--',linewidth=0.7,alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
ax = sns.countplot(data=gym, x='days_per_week')

ax.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=0.7)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
