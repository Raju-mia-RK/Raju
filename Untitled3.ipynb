find max price from 20 items

```python
arr=[]
print("Enter 20 item price")
for i in range (0,20):
  price=int(input())
  arr.append(price)
max=arr[0]
for i in range(1,20):
  if arr[i]>max:
    max=arr[i]

print("maximum price is ",max)
```

find sum of odd number

```python
print("Enter upper limit: ")
n=int (input())
sum=0.0
for i in range(1,n+1):
  if(i%2!=0):
    sum+=i

print("sum of odd number: ",sum)
```

factorial of a number

```python
print("Enter a number: ")
n=int(input())
fact=1
for i in range(1,n+1):
  fact*=i
print("Factorial value is: ",fact)
```
