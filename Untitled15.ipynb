```python
students = pd.read_csv('/content/drive/MyDrive/datasets/StudentsPerformance.csv')
```

```python
sns.histplot(students['math score'])
plt.title("Math score with seaborn")
plt.show()

sns.histplot(students['math score'],kde=True)
plt.show()
```

```python
score=students['math score']
plt.hist(score.values,label=score.index)

plt.show()
```

```python
sns.boxplot(data=students,x='gender',y='math score')
plt.grid(True,which='both',linestyle='--',linewidth=0.7,alpha=0.7)
plt.title("Math by gender")
plt.show()

sns.boxplot(data=students,x='lunch',y='math score')
plt.grid(True,which='both',linestyle='--',linewidth=0.7,alpha=0.7)
plt.title("Lunch by gender")
plt.show()
```

```python
print(students.columns.tolist())
```

```python
sns.boxplot(data=students,x='gender',y='parental level of education')
plt.grid(True,which='both',linestyle='--',linewidth=0.7,alpha=0.7)
plt.title("Parental level of education by Gender")
plt.show()
```

```python
sns.countplot(data=students,x='race/ethnicity')
plt.title("Race or ethnicity count")
plt.grid(True,which='both',linestyle='--',linewidth=0.7,alpha=0.7)
plt.show()
```

```python
sns.countplot(data=students,x='test preparation course')
plt.grid(True,which='both',linestyle='--',linewidth=0.7,alpha=0.7)
plt.show()
```

```python
sns.barplot(data=students,x='gender',y='math score')
plt.title("Average math score with gender")
plt.grid(True,which='both',linestyle='--',linewidth=0.7,alpha=0.7)

plt.show()
```

```python
sns.barplot(data=students,x='parental level of education',y='math score')
#plt.grid(True,which='both',linestle='--',linewidth=0.7,alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.title("Average parental level of education by math score")
plt.show()
```

```python
sns.kdeplot(data=students,x='math score',hue='gender')
plt.grid(True,which='both',linestyle='--',linewidth=0.7,alpha=0.7)
plt.title("Math score by gender")
plt.show()
```

```python
sns.kdeplot(data=students,x='writing score',hue='gender')
plt.grid(True,which='both',linestyle='--',linewidth=0.7,alpha=0.7)
plt.title("writing score by gender")
plt.show()
```

```python
sns.countplot(data=students,x='test preparation course')
plt.grid(True,which='both',linestyle='--',linewidth=0.7,alpha=0.7)
plt.show()
```

```python
scores=students[['math score','reading score','writing score']]
corrleation=scores.corr()
print(corrleation)
```

```python
sns.heatmap(corrleation)
plt.title("Scores corrleation")
plt.show()
```

```python
sns.heatmap(corrleation,annot=True)
plt.title("Scores Corrleation")
plt.show()
```

```python
sns.pairplot(scores)
plt.show()
```

```python
score_data=students[['math score','reading score','writing score','gender']]
sns.pairplot(score_data,hue='gender')
plt.show()
```
