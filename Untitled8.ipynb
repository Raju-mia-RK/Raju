```python
#Write a Python program that asks the user to input their age and prints whether they are eligible to vote (18 years or older) or not.
def vote(age):
  if(age>=18):
    print("Eligible to vote.")
  else:
    print("Not eligible to vote.")

age=int(input("Enter a number: "))
vote(age)
```

```python
#Write a Python program that asks for the user's age and prints the appropriate message:
#If the user is 18 or older, print 'Adult'
#If the user is 13 or older but under 18, print 'Teenager'
#If the user is younger than 13, print 'Child'
def message(age):
  if(age>=18):
    print("Adult")
  elif(age>=13):
    print("Teenager.")
  else:
    print("child")
age=int(input("Enter a age: "))
message(age)
```

```python
#Write a Python program that checks if a user’s input number is positive, negative, or zero.
#If the number is positive, print 'Positive Number'
#If the number is negative, print 'Negative Number'
#If the number is zero, print 'Zero'
def check_number(number):
  if(number==0):
    print("Zero")
  elif(number>0):
    print("Positive")
  else:
    print("Negative")
number=int(input("Enter a number: "))
check_number(number)
```

```python
#Use a ternary operator to determine if a number entered by the user is even or odd. The program should print 'Even Number' or 'Odd Number' based on the input.
number=int(input("Enter a number: "))
message="Even number" if number%2==0 else "Odd number"
print(message)
```

```python
#Write a Python program that takes the user’s age and prints which age group they belong to:
#'Child' for age 0–12
#'Teenager' for age 13–19
#'Adult' for age 20–64
#'Senior' for age 65 and above
def message(age):
  if(age>=0 and age<=12):
    print("Child")
  elif(age>=13 and age<=19):
    print("Teenager")
  elif(age>=20 and age<=64):
    print("Adult")
  elif(age>=65):
    print("Senior")
  else:
    print("Not a valid age")
age=int(input("Enter your age: "))
message(age)
```

```python
#Write a Python function sum_of_natural_numbers(n) that returns the sum of all natural numbers up to n. Use a loop inside the function.
def sum_of_natural_numbers(n):
  sum=0
  for i in range(0,n):
    sum+=i
  return sum

number=int(input("Enter a number: "))
result=sum_of_natural_numbers(number)
print(f"Summeation of  {number} is: {result}")
```

```python
#Write a Python function factorial(n) that calculates and returns the factorial of n. Use a loop to calculate the factorial.
def factorial(n):
  fact=1
  for i in range(1,n+1):
    fact*=i
  return fact

number=int(input("Enter a number: "))
result=factorial(number)
print(f"factorial of {number} is: {result}")
```

```python
#Write a Python function multiplication_table(n) that prints the multiplication table of n up to 10.
def multiplication_table(n):
  print(f"Multiplication table of {n} is: ")
  for i in range(1,11):
    print(f"{n}*{i}={n*i}")
number=int(input("Enter a number: "))
multiplication_table(number)
```
