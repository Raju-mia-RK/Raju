```python
plt.plot(students['math score'].head())
plt.title("Math Scores of  Students.")
plt.xlabel("Student Number")
plt.ylabel('Math Score')
plt.show()
```

```python
plt.plot(range(1,11),students['reading score'].head(10),marker='o')
plt.title("Student Reading Score")
plt.xlabel("Student number")
plt.ylabel("Reading Score")
plt.show()
```

```python
gender_count=students['gender'].value_counts()

plt.bar(['Female','Male'],gender_count.values,color=['red','green'])
plt.title("Gender count")
plt.show()
```

```python
avg_math=students.groupby('gender')['math score'].mean()

plt.grid(True,which='both',linestyle='--',linewidth=0.7,alpha=0.7)
plt.bar(avg_math.index,avg_math.values,color=['lightblue','salmon'])
plt.title("Average Math score by gender")
plt.ylabel("Math score")
plt.show()
```

```python
avg_reading=students.groupby('gender')['reading score'].mean()

plt.grid(True,which='both',linestyle='--',linewidth=0.7,alpha=0.7)
plt.title('Reading Score by gender')
plt.ylabel("Reading Score")
plt.bar(avg_reading.index,avg_reading.values,color=['pink','red'])
plt.show()
```

```python
from matplotlib.lines import lineStyles
plt.scatter(students['math score'],students['reading score'])
plt.grid(True,which='both',linewidth=0.7,alpha=0.7)
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.show()
```

```python
sample=students.head(100)
plt.scatter(sample['math score'],sample['reading score'],alpha=0.3)
plt.xlabel("Math Score")
plt.ylabel("reading Score")
plt.show()
```

```python
plt.hist(students['math score'],bins=20,color='green')
plt.ylabel('Math score')
plt.xlabel("Student number")
plt.show()
```

```python
test_prep=students['test preparation course'].value_counts()

plt.title("Test Preparation course(%)")
plt.pie(test_prep,labels=test_prep.index,autopct='%1.1f%%')
plt.show()

```

```python
education_count=students['parental level of education'].value_counts()
plt.title("Parental level of education")
plt.pie(education_count,labels=education_count.index,autopct='%1.1f%%')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1: Histogram
axes[0, 0].hist(students['math score'], bins=15)
axes[0, 0].set_title("Math Score")

# Plot 2: Bar chart
gender_count = students['gender'].value_counts()
axes[0, 1].bar(gender_count.index, gender_count.values)
axes[0, 1].set_title("Gender")

# Plot 3: Scatter chart
score = students.head(50)
axes[1, 0].scatter(score['math score'], score['reading score'], alpha=0.5)
axes[1, 0].set_title("Math vs Reading Score")

# Plot 4: Line graph
sample = students.head(30)
axes[1, 1].plot(sample['writing score'])
axes[1, 1].set_title("Writing Scores")

plt.tight_layout()
plt.show()
```
